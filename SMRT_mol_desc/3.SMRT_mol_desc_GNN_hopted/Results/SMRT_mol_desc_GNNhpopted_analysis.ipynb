```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = "//home/user/YixiTFG/TFG_Yixi/SMRT_mol_desc/3.SMRT_mol_desc_GNN_hopted/Results/Results_GNN_hpopted.tsv"
df = pd.read_csv (datafile, sep='\t')
```

```python
mol_weight = df ["mol_weight"]
diff = df ["diff"]
plt.scatter (mol_weight, diff)
plt.xlabel ("molecular_weight (g/mol)")
plt.ylabel ("diff")
plt.title ("diff vs molecular_weight")
plt.show ()
```

```python
xlogp = df ["xlogp"]
plt.scatter (xlogp, diff)
plt.xlabel ("xlogp")
plt.ylabel ("diff")
plt.title ("diff vs xlogp")
plt.show ()
```
